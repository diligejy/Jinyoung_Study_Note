# 2.1 튜토리얼 — 연속 선형 계획법 (Continuous Linear Programming)

## 기본 설정 (Set Up)

연속 선형 계획법(CLP, Continuous Linear Programming)에서 <strong>목적 변수(objective variable)</strong> $z$는 임의의 실수값을 가질 수 있습니다:

$$z \in \mathbb{R}$$

$z$의 값은 <strong>결정 변수(decision variables)</strong> $x_j$들의 함수입니다. 여기서 $j$는 $1$부터 $n$까지의 정수 인덱스입니다:

$$z = f(x_1, x_2, \ldots, x_n)$$

$$x_1, x_2, \ldots, x_n \in \mathbb{R}$$

총 $n$개의 <strong>연속형(continuous)</strong> 결정 변수가 존재합니다. CLP에서 함수 $f$는 <strong>선형 함수(linear function)</strong>입니다. 목표는 $z$의 <strong>최적값(optimal value)</strong>을 구하는 것으로, 최대화(예: 이익 극대화) 또는 최소화(예: 비용 최소화) 중 하나입니다:

$$\max \text{ 또는 } \min\; z = f(x_1, x_2, \ldots, x_n) = c_1 \cdot x_1 + c_2 \cdot x_2 + \ldots + c_n \cdot x_n$$

목적 함수는 결정 변수들의 선형 함수인 <strong>제약 조건(constraints)</strong> 집합을 만족해야 합니다. 결정 변수와 계수 $a_{ij}$의 합이 계수 $b_j$보다 작거나, 크거나, 같아야 합니다:

$$\begin{aligned}
&s.t. \\
&a_{11} \cdot x_1 + a_{12} \cdot x_2 + \ldots + a_{1n} \cdot x_n \leq b_1 \\
&a_{21} \cdot x_1 + a_{22} \cdot x_2 + \ldots + a_{2n} \cdot x_n \leq b_2 \\
&\vdots \\
&a_{m1} \cdot x_1 + a_{m2} \cdot x_2 + \ldots + a_{mn} \cdot x_n \leq b_m
\end{aligned}$$

> **참고:** 계수 $a$는 두 개의 하첨자를 가집니다. 두 번째 첨자는 대응하는 결정 변수의 인덱스이고, 첫 번째 첨자는 $b$의 인덱스에 해당합니다. 인덱스 $i$는 $1$부터 $m$까지이므로 최적화 함수는 총 $m$개의 제약 조건을 가집니다. 부등식 왼쪽 항을 <strong>좌변(LHS, Left Hand Side)</strong>, 오른쪽 항을 <strong>우변(RHS, Right Hand Side)</strong>이라 합니다.

---

## 압축형 표기 (Compact / Matrix Form)

합 연산자(summation)를 사용하면 목적 함수는 다음과 같이 표현됩니다:

$$\max \text{ 또는 } \min\; z = \sum_{j=1}^{n} c_j \cdot x_j = c \cdot x$$

여기서:

$$x = [x_1, x_2, \ldots, x_n]^T, \quad c = [c_1, c_2, \ldots, c_n]^T$$

행렬 $A$와 결정 변수 벡터의 곱으로 모든 제약 조건을 표현하면:

$$\begin{aligned}
&\max \text{ 또는 } \min\; z = \sum_{j=1}^{n} c_j \cdot x_j \\
&s.t. \\
&\sum_{j=1}^{n} a_{1j} \cdot x_j \leq b_1 \\
&\sum_{j=1}^{n} a_{2j} \cdot x_j \leq b_2
\end{aligned}$$

---

## 수요 제약 조건 (Demand Constraints)

위에서 언급했듯이, 수요(demand)에 관한 요구사항이 있을 수 있습니다. 제품 $j$의 수요를 $d_j$로 표기하면, 수요 제약 조건은 다음과 같은 형태가 됩니다:

$$x_1 \geq d_1$$
$$x_2 \geq d_2$$
$$\vdots$$
$$x_n \geq d_n$$

이를 압축형으로 표현하면:

$$x_j \geq d_j \quad \forall j \in [1, \ldots, n]$$

---

## 가용성 제약 조건 (Availability Constraints)

각 제품을 생산하는 데 여러 종류의 <strong>제한된 자원(resource)</strong>이 사용됩니다. $m$가지의 서로 다른 자원이 있다고 가정합니다. 자원 $i$ ($i \in [1, \ldots, m]$)의 가용량을 $b_i$로 표기합니다:

- $b_i$: 자원 $i$의 가용량 (자원 단위로 표현)

또한, 제품 $j$ 한 단위를 생산하는 데 필요한 자원 $i$의 소요량을 $a_{ij}$로 표기합니다:

- $a_{ij}$: 제품 $j$ 한 단위 생산에 필요한 자원 $i$의 소요량 (자원 단위 / 제품 단위)

각 자원 $i$에 대해, **사용된 총량이 가용량을 초과하지 않아야** 합니다. 특정 수량의 제품을 생산하는 데 소비되는 자원의 양은 생산 단위 수(예: $x_j$)에 단위당 소요량을 곱한 값입니다. 이를 모든 제품에 대해 합산한 결과가 가용량 이하여야 합니다.

자원 인덱스 $1$을 예로 들면:

$$x_1 \cdot a_{11} + x_2 \cdot a_{12} + \ldots + x_n \cdot a_{1n} \leq b_1$$

압축형으로 표현하면:

$$\sum_{j=1}^{n} x_j \cdot a_{1j} \leq b_1$$

나머지 자원들도 동일하게 적용합니다:

$$\sum_{j=1}^{n} x_j \cdot a_{2j} \leq b_2$$
$$\sum_{j=1}^{n} x_j \cdot a_{3j} \leq b_3$$
$$\vdots$$
$$\sum_{j=1}^{n} x_j \cdot a_{mj} \leq b_m$$

이 $m$개의 부등식을 압축형으로 표현하면:

$$\sum_{j=1}^{n} x_j \cdot a_{ij} \leq b_i \quad \forall i \in [1, \ldots, m]$$

즉, 각 자원에 대해 **모든 제품 유형을 생산하는 데 사용된 양의 합이 총 가용량 이하**여야 합니다.

---

## 요약 — 생산 믹스 문제의 CLP 공식화 (Summary)

지금까지의 내용을 종합하면, 생산 믹스 문제(Production Mix Problem)의 일반적인 CLP 공식화는 다음과 같습니다:

$$\max\; z = \sum_{j=1}^{n} x_j \cdot p_j$$

$$s.t.$$

$$\sum_{j=1}^{n} x_j \cdot a_{ij} \leq b_i \quad \forall i \in [1, \ldots, m]$$

$$x_j \geq d_j \quad \forall j \in [1, \ldots, n]$$

여기서 $p_j$는 제품 $j$의 단위당 이익(profit)입니다.

## 블렌딩 문제 (Blending Problems)

블렌딩 문제에서 의사결정자는 주어진 <strong>수요(demand)</strong>를 충족시키는 <strong>재료(materials) 또는 자원(resources)의 최적 혼합(blending)</strong>을 찾아야 합니다. 각 자원 또는 재료는 서로 다른 비용을 가지므로, 기본적인 형태에서 목적은 **수요를 충족하는 데 필요한 재료의 전체 비용을 최소화**하는 것입니다. 또한, 필요한 재료나 자원은 <strong>가용성 제약(availability constraints)</strong>을 받을 수 있습니다(즉, 재료나 자원의 가용량이 무한하다고 가정할 수 없습니다).

---

### 목적 함수 (Objective Function)

$m$가지의 서로 다른 재료 또는 자원, $n$가지의 서로 다른 제품이 있다고 가정합니다.

- $c_i$: 재료(material/resource) $i$의 비용, $i \in [1, \ldots, m]$
- $x_{ij}$: 제품 $j$를 생산하기 위해 사용되는 재료 $i$의 양, $j \in [1, \ldots, n]$

> **참고:** $x_{ij}$가 **결정 변수**입니다. 우리는 제품을 생산하기 위한 자원과 재료의 최적 조합을 찾고 있습니다. 결정 변수는 총 $m \times n$개입니다(재료와 제품의 조합).

이 정의들로부터 목적 함수는 다음과 같이 정의됩니다:

$$\min\; z = \sum_{i=1}^{m} \sum_{j=1}^{n} c_i \cdot x_{ij}$$

즉, 목표는 **모든 제품을 생산하는 데 필요한 서로 다른 재료들의 비용의 합인 전체 비용을 최소화**하는 것입니다.

---

### 가용성 제약 조건 (Availability Constraints)

자원이나 재료의 양이 제한되어 있으므로, 가용성 제약 조건을 설정합니다. 재료 $i$의 가용량을 $a_i$로 표기합니다. 모든 제품을 생산하는 데 사용된 총량은 해당 재료의 가용량을 초과할 수 없습니다.

예를 들어, 재료 1에 대해:

$$\sum_{j=1}^{n} x_{1j} \leq a_1$$

다른 재료들에도 동일하게 적용됩니다:

$$\sum_{j=1}^{n} x_{2j} \leq a_2$$

$$\vdots$$

$$\sum_{j=1}^{n} x_{mj} \leq a_m$$

압축형으로 표현하면:

$$\sum_{j=1}^{n} x_{ij} \leq a_i \quad i \in [1, \ldots, m]$$

---

### 수요 제약 조건 (Demand Constraints)

모든 제품에 대한 수요를 충족해야 합니다. 블렌딩 문제에서는 **해당 제품을 생산하는 데 사용된 모든 재료의 총량**이 제품의 총 생산량과 동일합니다. 제품 $j$의 수요를 $d_j$로 표기하면:

$$\sum_{i=1}^{m} x_{i1} \geq d_1$$

$$\sum_{i=1}^{m} x_{i2} \geq d_2$$

$$\vdots$$

$$\sum_{i=1}^{m} x_{in} \geq d_n$$

압축형으로 표현하면:

$$\sum_{i=1}^{m} x_{ij} \geq d_j \quad j \in [1, \ldots, n]$$

---

### 추가 요구 사항: 최소 구성 비율 (Additional Requirements: Minimum Amount)

문제 정의에서 특정 재료의 <strong>최소 구성 비율(minimum amount)</strong> 요구가 있을 수 있습니다. 이 요구는 제품의 전체 구성 중 특정 재료가 차지하는 <strong>비율(percentage/ratio)</strong>로 표현되기도 합니다.

예시로, 제품 3에 사용되는 재료 2의 최소 비율이 20%(0.2)여야 한다고 정의합니다. 즉, 재료 2의 사용량 $x_{23}$은 제품 3을 생산하는 데 사용된 모든 재료 합의 **최소 20% 이상**이어야 합니다:

$$x_{23} \geq 0.2 \cdot \sum_{i=1}^{m} x_{i3}$$

> **수식 검증 (원본 오류 수정):** 원본에는 $x_{23} \leq 0.2 \cdot \sum x_{i3}$로 표기되어 있으나, "최소(at least) 20%"라는 조건이므로 부등호 방향이 $\geq$가 맞습니다. 이하 유도도 이에 맞춰 수정합니다.

좌변으로 종속 계수를 이항하면:

$$x_{23} - 0.2 \cdot \sum_{i=1}^{m} x_{i3} \geq 0$$

$\sum_{i=1}^{m} x_{i3} = x_{23} + \sum_{i=1,\, i \neq 2}^{m} x_{i3}$를 대입하면:

$$x_{23} - 0.2 \cdot \left(x_{23} + \sum_{i=1,\, i \neq 2}^{m} x_{i3}\right) \geq 0$$

$$0.8 \cdot x_{23} - 0.2 \cdot \sum_{i=1,\, i \neq 2}^{m} x_{i3} \geq 0$$

## 생산 믹스와 블렌딩 문제의 결합 (Combination of Production Mix and Blending Problems)

전체 비용을 최소화하는 것 대신, **이익(profit)을 최대화**하는 것이 목적이 될 수 있습니다. 이는 생산 믹스 문제와 블렌딩 문제를 결합한 형태입니다. $n$개 제품의 단위 판매 가격을 $[p_1, p_2, \ldots, p_n]$로 정의합니다.

제품 1에 집중하면, 이익은 판매 가격과 생산에 소모된 재료 비용의 차이입니다. 모든 재료의 총량이 제품의 총 생산량과 동일하다고 가정하면, 제품 1의 전체 이익은 다음과 같습니다:

$$p_1 \cdot \sum_{i=1}^{m} x_{i1} - \sum_{i=1}^{m} c_i \cdot x_{i1}$$

> **수식 검증 (원본 오류 수정):** 원본에는 $p_1 \cdot \sum_{j=1}^{n} x_{1j}$로 표기되어 있으나, 제품 1의 총 생산량은 재료 인덱스 $i$를 합산해야 하므로 $\sum_{i=1}^{m} x_{i1}$이 맞습니다.

전체 이익을 최대화하는 목적 함수:

$$\max\; z = \sum_{j=1}^{n} p_j \cdot \sum_{i=1}^{m} x_{ij} - \sum_{i=1}^{m} c_i \cdot x_{ij}$$

합산 항을 재정리하면:

$$\max\; z = \sum_{j=1}^{n} \sum_{i=1}^{m} (p_j - c_i) \cdot x_{ij}$$

## 표준형 (Standard Form)

Simplex와 같은 일부 알고리즘은 <strong>표준형(standard form)</strong>으로 정식화된 문제를 필요로 합니다. 기본 설정 섹션에서 설명된 모든 CLP 문제는 다음 조건을 만족하는 표준형으로 변환할 수 있습니다:

- 목적 함수는 <strong>최대화(maximise)</strong> 형태
- 모든 제약 조건은 <strong>등식(equal)</strong> 형태

즉, 모든 CLP 문제는 다음 표준형으로 표현할 수 있습니다:

$$\max\; z = \sum_{i=1}^{n} c_i \cdot x_i = c^T x$$

$$s.t.$$

$$a_{11} \cdot x_1 + a_{12} \cdot x_2 + \ldots + a_{1n} \cdot x_n + s_1 = b_1$$
$$a_{21} \cdot x_1 + a_{22} \cdot x_2 + \ldots + a_{2n} \cdot x_n + s_2 = b_2$$
$$\vdots$$
$$a_{m1} \cdot x_1 + a_{m2} \cdot x_2 + \ldots + a_{mn} \cdot x_n + s_m = b_m$$

이를 행렬 곱으로 표현하면:

$$A \cdot x = b$$

그리고:

$$x_i \geq 0 \quad \forall i$$

---

### 최소화 문제의 변환 (Conversion of Minimization Problems)

목적 함수가 최소화 형태인 경우, 다음 변수 치환(change of variable)만으로 표준형으로 변환할 수 있습니다:

$$z = -z^*$$

이 변수 치환으로 목적 함수:

$$\min\; z = c \cdot x$$

는 다음과 동치입니다:

$$\max\; z^* = -c \cdot x$$

이로써 목적 함수가 최대화 형태가 됩니다.

---

### 제약 조건 변환과 슬랙 변수 (Conversion of Constraints and Slack Variables)

등식 형태로 변환하기 위해 새로운 결정 변수 집합 $s = [s_1, s_2, \ldots, s_n]$을 도입합니다. 예를 들어, 문제에 다음과 같은 $\leq$ 형태의 제약 조건이 있다고 합니다:

$$a_{11} \cdot x_1 + a_{12} \cdot x_2 + \ldots + a_{1n} \cdot x_n \leq b_1$$

좌변이 우변보다 작거나 같으므로, 비음수 변수 $s_1$을 더해 등식으로 만들 수 있습니다:

$$a_{11} \cdot x_1 + a_{12} \cdot x_2 + \ldots + a_{1n} \cdot x_n + s_1 = b_1$$

두 제약 조건이 동치가 되려면 $s_1$에 비음성 제약이 필요합니다:

$$s_1 \geq 0$$

반대로, $\geq$ 형태의 제약 조건이 있으면 동일한 방법으로 변환할 수 있지만, 이때 새 변수는 문제를 동치로 만들기 위해 **0 이하**여야 합니다. 예를 들어:

$$a_{21} \cdot x_1 + a_{22} \cdot x_2 + \ldots + a_{2n} \cdot x_n \geq b_2$$

이는 다음과 동치입니다:

$$a_{21} \cdot x_1 + a_{22} \cdot x_2 + \ldots + a_{2n} \cdot x_n + s_2 = b_2$$

단:

$$s_2 \leq 0$$

> **수식 검증 (원본 오류 수정):** 원본에 $s_1 \leq 0$으로 표기되어 있으나, 두 번째 제약의 슬랙 변수이므로 $s_2 \leq 0$이 맞습니다.

또한, 등식 형태의 제약 조건인 경우 변경이 필요 없지만, 일관성을 위해 새 변수를 도입합니다:

$$a_{31} \cdot x_1 + a_{32} \cdot x_2 + \ldots + a_{3n} \cdot x_n = b_3$$

이는 다음과 동치입니다:

$$a_{31} \cdot x_1 + a_{32} \cdot x_2 + \ldots + a_{3n} \cdot x_n + s_3 = b_3$$

단:

$$s_3 = 0$$

> **수식 검증 (원본 오류 수정):** 원본에 $s_1 = 0$으로 표기되어 있으나, 세 번째 제약의 슬랙 변수이므로 $s_3 = 0$이 맞습니다.

---

### 슬랙 변수 (Slack Variables)

이 새로운 변수들을 <strong>슬랙 변수(slack variables)</strong>라고 하며, 좌변(LHS)과 우변(RHS)의 차이를 나타냅니다.

- <strong>잉여 변수(surplus variables)</strong>: 슬랙 변수가 비음수($s \geq 0$)일 때, 즉 $\leq$ 형태의 제약인 경우. 이 유형의 제약 조건은 일반적으로 제한된 자원을 나타내므로, 슬랙 값은 최적해에서 보유한 <strong>자원의 잉여량(surplus)</strong>을 나타냅니다.

- <strong>초과 변수(excess variables)</strong>: 슬랙 변수가 비양수($s \leq 0$)일 때, 즉 $\geq$ 형태의 제약인 경우. 이 유형의 제약 조건은 일반적으로 추가적인 요구 사항을 나타내므로, 슬랙 값은 요구 사항 대비 해의 <strong>초과량(excess)</strong>을 나타냅니다.

- <strong>인공 변수(artificial variables)</strong>: 슬랙 변수가 0일 때($s = 0$), 즉 $=$ 형태의 제약인 경우. 값이 반드시 0이어야 하므로, Simplex 같은 솔버에서 별도로 처리되어야 합니다.

### 음수 결정 변수의 변환 (Conversion of Negative Decision Variables)

표준형에서 모든 결정 변수는 비음수여야 합니다. 다음 변수 치환으로 이를 달성할 수 있습니다:

$$x = x^{+} - x^{++}$$

$$x^{+},\; x^{++} \in [0, \infty]$$

즉, 결정 변수 $x$를 두 개의 비음수 결정 변수로 대체합니다. $x^{++}$가 $x^{+}$보다 클 때, $x$는 음수 값을 가집니다.

### 요약 — 표준형 (Summary: Standard Form)

표준형이 성립하려면, 새로운 슬랙 변수를 결정 변수 벡터에 추가해야 합니다:

$$x = [x_1, x_2, \ldots, x_n, s_1, s_2, \ldots, s_p, a_1, a_2, \ldots, a_k]$$

여기서 $p$개의 비제로 슬랙 변수와 $k$개의 인공 변수가 있습니다.

목적 함수에서 인공 변수는 0이 되어야 하므로 임의의 큰 음수 $-M$을 곱합니다. 비제로 슬랙 변수는 0 계수를 가집니다. 따라서 목적 함수는:

$$\max\; z = \sum_{i=1}^{n} c_i \cdot x_i + 0 \cdot s_1 + 0 \cdot s_2 + \ldots + 0 \cdot s_p - M \cdot a_1 - M \cdot a_2 - \ldots - M \cdot a_k$$

제약 조건(확장 후):

$$a_{11} \cdot x_1 + a_{12} \cdot x_2 + \ldots + a_{1n} \cdot x_n + s_1 + 0 \cdot s_2 + \ldots + 0 \cdot s_p + 0 \cdot a_1 + 0 \cdot a_2 + \ldots + 0 \cdot a_k = b_1$$
$$a_{21} \cdot x_1 + a_{22} \cdot x_2 + \ldots + a_{2n} \cdot x_n + 0 \cdot s_1 + s_2 + \ldots + 0 \cdot s_p + 0 \cdot a_1 + 0 \cdot a_2 + \ldots + 0 \cdot a_k = b_2$$
$$\vdots$$
$$a_{p1} \cdot x_1 + a_{p2} \cdot x_2 + \ldots + a_{pn} \cdot x_n + 0 \cdot s_1 + 0 \cdot s_2 + \ldots + s_p + 0 \cdot a_1 + 0 \cdot a_2 + \ldots + 0 \cdot a_k = b_p$$
$$a_{(p+1)1} \cdot x_1 + a_{(p+1)2} \cdot x_2 + \ldots + a_{(p+1)n} \cdot x_n + 0 \cdot s_1 + 0 \cdot s_2 + \ldots + 0 \cdot s_p + a_1 + 0 \cdot a_2 + \ldots + 0 \cdot a_k = b_{p+1}$$
$$a_{(p+2)1} \cdot x_1 + a_{(p+2)2} \cdot x_2 + \ldots + a_{(p+2)n} \cdot x_n + 0 \cdot s_1 + 0 \cdot s_2 + \ldots + 0 \cdot s_p + 0 \cdot a_1 + a_2 + \ldots + 0 \cdot a_k = b_{p+2}$$
$$\vdots$$
$$a_{m1} \cdot x_1 + a_{m2} \cdot x_2 + \ldots + a_{mn} \cdot x_n + 0 \cdot s_1 + 0 \cdot s_2 + \ldots + 0 \cdot s_p + 0 \cdot a_1 + 0 \cdot a_2 + \ldots + 1 \cdot a_k = b_m$$

### 표준 표형 표현 (Standard Tabular Representation)

이 섹션의 시작에서 언급했듯이, 표준형을 다음과 같이 표현할 수 있습니다:

$$\max\; z = c \cdot x$$

$$s.t. \quad A \cdot x = b$$

여기서:

$$x = [x_1, x_2, \ldots, x_n, s_1, s_2, \ldots, s_p, a_1, a_2, \ldots, a_k]^T$$

$$c = [c_1, c_2, \ldots, c_n, 0, \ldots, 0, -M, \ldots, -M]^T$$

$$A = \begin{bmatrix}
a_{11} & a_{12} & \cdots & a_{1n} & 1 & 0 & \cdots & 0 \\
a_{21} & a_{22} & \cdots & a_{2n} & 0 & 1 & \cdots & 0 \\
\vdots & \vdots & \ddots & \vdots & \vdots & \vdots & \ddots & \vdots \\
a_{m1} & a_{m2} & \cdots & a_{mn} & 0 & 0 & \cdots & 1
\end{bmatrix}$$

$x$와 $c$는 $n + m$ 벡터이며, $A$는 $m \times (n+m)$ 행렬입니다. 오른쪽 $n$열은 원본 LHS 계수를, 맨 오른쪽 $m$열은 $n \times n$ 단위 행렬을 나타냅니다.

모든 종속 항을 LHS로 이항하면 목적 함수는:

$$z - c_1 \cdot x_1 - c_2 \cdot x_2 - \ldots - c_n \cdot x_n - 0 \cdot s_1 - 0 \cdot s_2 - \ldots - 0 \cdot s_p + M \cdot a_1 + M \cdot a_2 + \ldots + M \cdot a_k = 0$$

각 열이 변수의 계수를, 각 행이 방정식을 나타내는 데이터 프레임 형태로 모든 계수를 작성할 수 있습니다:

| $z$ | $x_1$ | $x_2$ | $\cdots$ | $x_n$ | $s_1$ | $s_2$ | $\cdots$ | $s_p$ | $a_1$ | $a_2$ | $\cdots$ | $a_k$ | RHS |
|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|
| 1 | $-c_1$ | $-c_2$ | $\cdots$ | $-c_n$ | 0 | 0 | $\cdots$ | 0 | $-M$ | $-M$ | $\cdots$ | $-M$ | 0 |
| 0 | $a_{11}$ | $a_{12}$ | $\cdots$ | $a_{1n}$ | 1 | 0 | $\cdots$ | 0 | 0 | 0 | $\cdots$ | 0 | $b_1$ |
| 0 | $a_{21}$ | $a_{22}$ | $\cdots$ | $a_{2n}$ | 0 | 1 | $\cdots$ | 0 | 0 | 0 | $\cdots$ | 0 | $b_2$ |
| $\vdots$ | $\vdots$ | $\vdots$ | $\ddots$ | $\vdots$ | $\vdots$ | $\vdots$ | $\ddots$ | $\vdots$ | $\vdots$ | $\vdots$ | $\ddots$ | $\vdots$ | $\vdots$ |
| 0 | $a_{m1}$ | $a_{m2}$ | $\cdots$ | $a_{mn}$ | 0 | 0 | $\cdots$ | 0 | 0 | 0 | $\cdots$ | 1 | $b_m$ |

이 표현을 <strong>표준 표형(standard tabular form)</strong>이라고 합니다.
