## 문제 모델 (Problem Model)

원본 문제: **Making Weapons** (무기 제조 최적화)

**목적 함수 (최대화):**

$$\max z = 40x_{1} + 5x_{2}$$

**결정 변수:**
- $x_1$: 무기 유형 1 생산량
- $x_2$: 무기 유형 2 생산량

**제약 조건:**

$$x_{1} + 0.5x_{2} \leq 125 \quad \text{(베스카 합금 제약)}$$

$$x_{1} + x_{2} \leq 225 \quad \text{(텅스텐 제약)}$$

$$3x_{1} + x_{2} \leq 300 \quad \text{(티타늄 제약)}$$

$$x_{2} \geq 100 \quad \text{(수요 제약)}$$

**사용법:** 아래 코드 셀을 실행하면 인터랙티브 슬라이더와 체크박스가 나타납니다.
- 슬라이더: $z$ 값 조정 → 목적 함수 등고선 이동
- 체크박스: 제약 조건 하나씩 추가 → 가능 영역 변화 관찰
- **최적값**: 목적 함수 선이 가능 영역(회색)과 접하는 최고 $z$ 지점

## 그래픽 해법 시뮬레이션 (Interactive Graphic Solution)

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interactive
matplotlib.rcParams["font.family"] = "Malgun Gothic"
matplotlib.rcParams["axes.unicode_minus"] = False

x = np.linspace(0, 300, 2000)

y1 = (125 - x) / 0.5  # 베스카 제약
y3 = 225 - x           # 텅스텐 제약
y4 = 100 + x * 0       # 수요 제약
y5 = 300 - 3 * x       # 티타늄 제약

def show_solution(z, beskar_constraint, tungsten_constraint, titanium_constraint, demand_constraint):
    y_min = x * 0
    y_max = 250 + 0 * x
    plt.figure(1)
    if beskar_constraint:
        plt.plot(x, y1, label=r'$x_{1}+0.5x_{2}\leq125$ (베스카)')
        y_max = np.minimum(y1, y_max)
    if tungsten_constraint:
        plt.plot(x, y3, label=r'$x_{1}+x_{2}\leq225$ (텅스텐)')
        y_max = np.minimum(y3, y_max)
    if titanium_constraint:
        plt.plot(x, y5, label=r'$3x_{1}+x_{2}\leq300$ (티타늄)')
        y_max = np.minimum(y5, y_max)
    if demand_constraint:
        plt.plot(x, y4, label=r'$x_{2}\geq100$ (수요)')
        y_min = y4
    plt.xlim((0, 150))
    plt.ylim((0, 250))
    plt.xlabel(r'$x_{1}$')
    plt.ylabel(r'$x_{2}$')
    plt.fill_between(x, y_max, y_min, where=y_max > y_min, color='grey', alpha=0.5)
    obj_func = (z - 40 * x) / 5
    plt.plot(x, obj_func, label=r'$z = 40x_1 + 5x_2$', linestyle='--', color='red')
    plt.title(f'그래픽 해법 시뮬레이션 (z = {z})')
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.show()

interactive_plot = interactive(
    show_solution,
    z=(0, 6000, 10),
    beskar_constraint=False,
    tungsten_constraint=False,
    titanium_constraint=False,
    demand_constraint=False
)
output = interactive_plot.children[-1]
output.layout.height = '400px'
interactive_plot
```

## 실험 가이드 (Experiment Guide)

**순서대로 체크박스를 켜면서 관찰하세요:**

1. **베스카 제약 ON** → 가능 영역 상단 경계 형성
2. **텅스텐 제약 ON** → 추가 제약으로 영역 축소
3. **티타늄 제약 ON** → 더 강한 제약 추가
4. **수요 제약 ON** → 하단 경계 형성 (최소 $x_2 = 100$ 보장)
5. 모든 제약 ON 후 → 슬라이더로 $z$를 올리며 **가능 영역과 접하는 최고 z 탐색**

**핵심 관찰:**
- 최적해는 항상 가능 영역의 <strong>꼭짓점(모서리)</strong>에 위치
- 목적 함수선이 가능 영역을 벗어나는 순간이 바로 **최적값을 초과**한 것

> 이 직관이 바로 다음 단원에서 배울 <strong>심플렉스 알고리즘(Simplex Algorithm)</strong>의 핵심 아이디어입니다.
