# 2.1 Tutorials — Continuous Linear Programming

## Set up

In CLP, the **objective variable**, noted as $z$, can take any real value:

$$z \in \mathbb{R}$$

The value of $z$ is a function of another set of unknowns called **decision variables**, noted as $x_j$, where $j$ is an integer index that goes from $1$ to $n$:

$$z = f(x_1, x_2, \ldots, x_n)$$

$$x_1, x_2, \ldots, x_n \in \mathbb{R}$$

We will have $n$ different **continuous** decision variables. In CLP, the function $f$ is a **linear function**. We want to find the **optimal value** of $z$ — either maximum (e.g. maximise profits) or minimum (e.g. minimise costs):

$$\max \text{ or } \min\; z = f(x_1, x_2, \ldots, x_n) = c_1 \cdot x_1 + c_2 \cdot x_2 + \ldots + c_n \cdot x_n$$

The objective function is subject to a set of **constraints**, which are also linear functions of the decision variables. The sum product of the decision variables times coefficients $a_{ij}$ must be less than, greater than, or equal to another coefficient $b_j$:

$$\begin{aligned}
&s.t. \\
&a_{11} \cdot x_1 + a_{12} \cdot x_2 + \ldots + a_{1n} \cdot x_n \leq b_1 \\
&a_{21} \cdot x_1 + a_{22} \cdot x_2 + \ldots + a_{2n} \cdot x_n \leq b_2 \\
&\vdots \\
&a_{m1} \cdot x_1 + a_{m2} \cdot x_2 + \ldots + a_{mn} \cdot x_n \leq b_m
\end{aligned}$$

```{note}
Coefficients $a$ have two sub-indexes: the second equals the index of the corresponding decision variable, the first equals the index of $b$. Index $i$ ranges from $1$ to $m$, so the optimisation function is subject to $m$ constraints. Expressions to the left of the inequality are the **Left Hand Side (LHS)**, and to the right are the **Right Hand Side (RHS)**.
```

## Compact (Matrix) Form

Using the sum operator, the objective function becomes:

$$\max \text{ or } \min\; z = \sum_{j=1}^{n} c_j \cdot x_j = c \cdot x$$

where:

$$x = [x_1, x_2, \ldots, x_n]^T, \quad c = [c_1, c_2, \ldots, c_n]^T$$

Expressing all constraints as the product of matrix $A$ and the decision variable vector:

$$\begin{aligned}
&\max \text{ or } \min\; z = \sum_{j=1}^{n} c_j \cdot x_j \\
&s.t. \\
&\sum_{j=1}^{n} a_{1j} \cdot x_j \leq b_1 \\
&\sum_{j=1}^{n} a_{2j} \cdot x_j \leq b_2
\end{aligned}$$
